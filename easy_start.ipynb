```python
import random
```

```python
class Chain:

    def __init__(self,id):
        self.id = id
        self.connected = []
        self.value = 0

    def set_chain_connected(self,chains):
        self.connected = chains
        
```

```python
class ChainGeneration:

    def __init__(self,chain_id):
        self.chain_id = chain_id
        self.chain = []
        self.chain_objs = []


    def generate_chain(self,n):
        arr = list(str(int(random.uniform(10**n,10**n*10))))
        self.chain = [int(i) for i in arr]

    def get_chain(self):
        return self.chain
    

    def get_chain_objects(self,objects):
        self.chain_objs = objects

    def apply_rules(self,chain):
        if chain.value == 0:
            chain.value +=1

        if sum(chain.connected) % 3 == 0:
            chain.value += 1

        values = random.choices(chain.connected,k=3)
        if sum(values) % 2 == 0:
            chain.value += 1
        else:
            chain.value -= 1

    def adapt_chain(self,memory=[]):
        memory.append(self.chain)
        for chain in self.chain_objs:
            self.apply_rules(chain)

        new_chain =  [chain.value for chain in self.chain_objs]
        if new_chain not in memory:
            self.adapt_chain(memory=memory)
        else:
            self.chain = new_chain

        

        
            

        
    

```

```python
chainGen = ChainGeneration(1)
chainGen.generate_chain(10)
chainSeq = chainGen.get_chain()


chainList = []

for i,ch in enumerate(chainSeq):
    arr = chainSeq.copy()
    del arr[i]
    chain = Chain(i)
    chain.set_chain_connected(arr)
    chainList.append(chain)

chainGen.get_chain_objects(chainList)
chainGen.chain
chainGen.adapt_chain()
```
